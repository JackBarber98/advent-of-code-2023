```python
INPUT_PATH = './input.txt'
```

```python
with open(INPUT_PATH, 'r') as input_file:
    input_array = [line.strip() for line in input_file.readlines()]
input_array
```

```python


formatted_input = input_array
numbers_as_strings = {'one': 1, 'two': 2, 'three': 3, 'four': 4, 'five': 5, 'six': 6, 'seven': 7, 'eight': 8, 'nine': 9}

for key, value in numbers_as_strings.items():
    for index, line in enumerate(input_array):
        if key in line:
            formatted_input[index] = line.replace(key, str(value))
print(formatted_input)
```

```python
for index, line in enumerate(formatted_input):
    numerical_format = ''.join(filter(str.isdigit, line))
    formatted_input[index] = str(numerical_format[0]) + str(numerical_format[-1])
print(formatted_input)
```
